## Data Preprocessing :

```python
# Drop the 'Unnamed: 0' column

train_df = train_df.drop(columns = ['Unnamed: 0'])
test_df = test_df.drop(columns = ['Unnamed: 0'])
```

```python
# Combine train and test data for preprocessing

df = pd.concat([train_df, test_df], axis = 0)
```

```python
# Encode categorical variables

label_encoders = {}
categorical_cols = df.select_dtypes(include = ['object']).columns

for col in categorical_cols:
    le = LabelEncoder()
    le.fit(df[col])
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# Split back into train and test data

train_df = df[:len(train_df)]
test_df = df[len(train_df):]
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# Split features and target variable

X_train = train_df.drop('is_fraud', axis = 1)
y_train = train_df['is_fraud']

X_test = test_df.drop('is_fraud', axis = 1)
y_test = test_df['is_fraud']
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
# Feature Scaling

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Exploratory Data Analysis :

```python
# Visualize the distribution of target variable

plt.figure(figsize = (6,4))
sns.countplot(x = 'is_fraud', data = train_df)
plt.title('Distribution of Fraudulent Transactions')
plt.show()
```

## Model Building and Evaluation :

```python
# Store accuracy scores

accuracy_scores = {}
```

```python
# Logistic Regression

lr = LogisticRegression(max_iter = 1000, class_weight = 'balanced')
lr.fit(X_train_scaled, y_train)
y_pred_lr_train = lr.predict(X_train_scaled)
y_pred_lr_test = lr.predict(X_test_scaled)

train_acc_lr = accuracy_score(y_train, y_pred_lr_train)
test_acc_lr = accuracy_score(y_test, y_pred_lr_test)
accuracy_scores['Logistic Regression'] = test_acc_lr

print("Logistic Regression :- ")
print(f"\nTraining Accuracy : {train_acc_lr}")
print(f"Testing Accuracy : {test_acc_lr}")
conf_matrix_lr = confusion_matrix(y_test, y_pred_lr_test)
print(conf_matrix_lr)
print(classification_report(y_test, y_pred_lr_test))
```

```python
# Decision Tree 

dt = DecisionTreeClassifier(class_weight = 'balanced')
dt.fit(X_train_scaled, y_train)
y_pred_dt_train = dt.predict(X_train_scaled)
y_pred_dt_test = dt.predict(X_test_scaled)

train_acc_dt = accuracy_score(y_train, y_pred_dt_train)
test_acc_dt = accuracy_score(y_test, y_pred_dt_test)
accuracy_scores['Decision Tree'] = test_acc_dt

print("Decision Tree :- ")
print(f"\nTraining Accuracy : {train_acc_dt}")
print(f"Testing Accuracy : {test_acc_dt}")
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt_test)
print(conf_matrix_dt)
print(classification_report(y_test, y_pred_dt_test))
```

```python
# Random Forest 

rf = RandomForestClassifier(class_weight = 'balanced')
rf.fit(X_train_scaled, y_train)
y_pred_rf_train = rf.predict(X_train_scaled)
y_pred_rf_test = rf.predict(X_test_scaled)

train_acc_rf = accuracy_score(y_train, y_pred_rf_train)
test_acc_rf = accuracy_score(y_test, y_pred_rf_test)
accuracy_scores['Random Forest'] = test_acc_rf

print("Random Forest :- ")
print(f"Training Accuracy : {train_acc_rf}")
print(f"Testing Accuracy : {test_acc_rf}")
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf_test)
print(conf_matrix_rf)
print(classification_report(y_test, y_pred_rf_test))
```
